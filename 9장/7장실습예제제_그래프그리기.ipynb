```python
import numpy as np
import matplotlib.pyplot as plt

gu = ['성산구','의창구','마산합포구','마산회원구','진해구']
length=np.zeros(5)

for i in range(5):
    length[i] = len(address[address.str.contains(gu[i])])

plt.bar(gu,length,color=('orange','r','b','m','c'))
plt.show()
```

```python
plt.pie(length, labels=gu)
plt.show()
```
