```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.plot(x, values)
plt.xlabel('years'), plt.ylabel('values'), plt.title('value per year')
plt.xticks(x, years)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values)
plt.xlabel('years'), plt.ylabel('values'), plt.title('value per year')
plt.xticks(x, years)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, color = 'dodgerblue') #dodgerblue, C2, #e35f62
plt.xlabel('years'), plt.ylabel('values'), plt.title('value per year')
plt.xticks(x, years)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10] #Banana랑 Grapes는 10%씩 떼어내라
#plt.pie(ratio)
#plt.pie(ratio, labels = labels, autopct = '%.1f%%')
plt.pie(ratio, labels = labels, autopct = '%.1f%%', startangle=260, 
        counterclock=False, explode = explode, shadow = True)
plt.show()
```
