```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018','2019','2020']
values = [100,400,900]

# 꺽은선 그래프
plt.plot(x, values)
plt.xlabel('years'), plt.ylabel('values'), plt.title('value per year')
plt.xticks(x, years)
plt.show()

# 근데 plt.plot(years,values )
# plt.xlabel('years'), plt.ylabel('values'), plt.title('value per year')
# 이것도 값 똑같이 나옴
```

```python
plt.bar(years, values)
plt.xlabel('years'), plt.ylabel('values'), plt.title('value per year')
plt.show()
```

```python
plt.bar(years, values, color='dodgerblue')  #dodgerblue, C2, #e35f62
plt.xlabel('years'), plt.ylabel('values'), plt.title('value per year')
plt.show()
```

```python
ratio = [34,32,16,18] # 비율
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10] #Banana랑 Grapes는 10%씩 떼어내라

plt.pie(ratio, labels = labels,  startangle=260,
        autopct='%.1f%%',shadow =  True, explode=explode, counterclock=False)
plt.show()
```

- autopct : 부채꼴 안에 표시될 숫자의 형식을 지정

autopct= 퍼센트를 파이차트 안에 표시   
autopct = '%.1f' 소수 첫째자리까지 표시  
autopct = '%.2f 소수 둘째자리까지 표시  
autopct = '%.1f%%' 퍼센트까지 포함  

- startangle는 부채꼴이 그려지는 시작 각도를 설정-.디폴트는 0도 (양의 방향 x축)로 설정  
- counterclock=False로 설정하면 시계 방향 순서로 부채꼴 영역이 표시
- shadow= True로 설정하면, 파이 차트에 그림자가 표시됩니다.
