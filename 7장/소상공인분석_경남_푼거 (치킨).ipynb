```python
import chardet
import pandas as pd
filename = "소상공인시장진흥공단_상가(상권)정보_경남_202312.csv"
with open(filename, 'rb') as f:
    result = chardet.detect(f.readline())  # or read() if the file is small.
    print(result['encoding'])
```

```python
data = pd.read_csv('소상공인시장진흥공단_상가(상권)정보_경남_202312.csv',
                  encoding='utf-8')
data
```

```python
# 창원시 포함된 곳 출력하라
changwon = data[data['도로명주소'].str.contains('창원시')]
changwon
```

```python
# 상호명이 치킨인곳 골라라
chicken =changwon[changwon['상호명'].str.contains('치킨')]
chicken
```
