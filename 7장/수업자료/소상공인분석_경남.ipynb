```python
import chardet
import pandas as pd
filename = '소상공인시장진흥공단_상가(상권)정보_경남_202312.csv'
with open(filename,'rb') as f:
    result = chardet.detect(f.readline())
    print(result['encoding'])
```

```python
data = pd.read_csv('소상공인시장진흥공단_상가(상권)정보_경남_202312.csv',
                  encoding='UTF-8')
data
```

```python
temp = data[['상호명','상권업종대분류명','시군구명','경도','위도']]
temp
```

```python
changwon = temp[temp['시군구명'].str.contains('창원시')]
changwon
```

```python
# 치킨집을 차리고싶은데 치킨집이 얼마나 많징!?
df = changwon[changwon['상호명'].str.contains('치킨')]
df
```

```python
# 상호명에 치킨만 들어잇겟냥?!
df = changwon[changwon['상호명'].str.contains('닭')]
df
```
