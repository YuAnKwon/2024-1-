```python
# groupby 사용법
# groupby() 괄호안에 있는게 index, 
# 컬럼은 agg뒤에 오는 계산값들에 의함.
# 어떤 컬럼? 그걸 [] 안에 지정해주고, 안해주면 숫자는 다 알아서 해줌
# tip_pct의 평균을 넣어줌

grouped = tips.groupby(['day','smoker'])
grouped[['tip_pct']].agg('mean')
```

```python
# 어떤 컬럼? 그걸 [] 안에 지정해주고, 안해주면 숫자는 다 알아서 해줌
# 안해주는 경우
grouped = tips.groupby(['day','smoker'])
grouped.agg('mean')
```

```python
#팁 데이터에서 그룹별 상위 5개의 tip_pct 값을 고른다.
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]


top(tips, n=6)
```

```python
frame = pd.DataFrame({'data1': np.random.randn(1000),
                     'data2': np.random.randn(1000)})
frame
```

```python
quartiles = pd.cut(frame.data1, 4)
quartiles
```

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
           'count': group.count(), 'mean': group.mean()}

grouped = frame.data2.groupby(quartiles)
grouped.apply(get_stats).unstack()
```

```python
import pandas as pd

# 데이터 읽기
tips = pd.read_csv('수업자료/tips.csv')

# 팁 비율 열 추가
tips['tip_pct'] = tips['tip'] / tips['total_bill']

# 피벗 테이블 생성
pivot = tips.pivot_table(
    index=['day', 'smoker'],
    values=['total_bill', 'tip', 'tip_pct'],
    aggfunc={'total_bill': 'mean', 'tip': 'mean', 'tip_pct': 'mean'}
)

pivot
```

```python
tips.pivot_table(index=['day','smoker'])
```
