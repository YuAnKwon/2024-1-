```python
import chardet
import pandas as pd
filename = '소상공인시장진흥공단_상가(상권)정보_경남_202312.csv'
with open(filename,'rb') as f:
    result = chardet.detect(f.readline())
    print(result['encoding'])
```

```python
temp = pd.read_csv('소상공인시장진흥공단_상가(상권)정보_경남_202312.csv',
                  encoding='utf-8')
temp
```

```python
changwon = temp[temp['시군구명'].str.contains('창원시')]

chicken = changwon[changwon['상호명'].str.contains('치킨')]
chicken = changwon[changwon['상호명'].str.contains('치킨|닭')]
```

1. df.DataFrame에서 '교촌치킨, ' BBQ치킨, 장모님치킨으로만 구성된 각각의 DataFrame을 만드시오
2. concat 방식을 써서 3개의 DataFrame을 합치시오

```python
Kchon = chicken[chicken['상호명'].str.contains('교촌')]
BBQ = chicken[chicken['상호명'].str.contains('BBQ')]
JangMo = chicken[chicken['상호명'].str.contains('장모님')]
```

```python
pd.concat([Kchon,BBQ,JangMo])
```
