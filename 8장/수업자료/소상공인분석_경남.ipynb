```python
import chardet
import pandas as pd
filename = '소상공인시장진흥공단_상가(상권)정보_경남_202312.csv'
with open(filename,'rb') as f:
    result = chardet.detect(f.readline())
    print(result['encoding'])
```

```python
data = pd.read_csv('소상공인시장진흥공단_상가(상권)정보_경남_202312.csv',
                  encoding='UTF-8')
data
```

```python
data.columns
```

```python
temp = data[['상호명','상권업종대분류명','시군구명','경도','위도']]
temp
```

```python
changwon = temp[temp['시군구명'].str.contains('창원시')]
changwon
```

```python
# 치킨집을 차리고싶은데 치킨집이 얼마나 많징!?
df = changwon[changwon['상호명'].str.contains('치킨')]
df
```

```python
# 상호명에 치킨만 들어잇겟냥?!
df = changwon[changwon['상호명'].str.contains('닭')]
df
```

```python
# 둘다 같이 !!!
df = changwon[changwon['상호명'].str.contains('닭|치킨')] # '닭'|'치킨'은 오류남
df
```

1. df.DataFrame에서 '교촌치킨, ' BBQ치킨, 장모님치킨으로만 구성된 각각의 DataFrame을 만드시오
2. concat 방식을 써서 3개의 DataFrame을 합치시오

```python
Kchon= df[df['상호명'].str.contains('교촌치킨')]
Kchon
```

```python
BBQ = df[df['상호명'].str.contains('비비큐|BBQ')]
BBQ
```

```python
jangmo= df[df['상호명'].str.contains('장모님치킨')]
jangmo
```

```python
len(BBQ)
```

```python
len(Kchon)
```

```python
len(jangmo)
```

```python
changwon_chicken = pd.concat([Kchon, BBQ, jangmo])
len(changwon_chicken)
```

```python
# 상권업종대분류명 없애고 인덱스 똑바로 !!
changwon_chicken.drop('상권업종대분류명', axis=1)
```
