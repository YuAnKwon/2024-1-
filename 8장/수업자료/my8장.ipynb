```python
import numpy as np
import pandas as pd
fruit = pd.DataFrame({'Num':[123, 456, 789, 1011, 1112],
                      'Fruit':['Apple', 'Banana', 'Cherry', 'Lemon', 'Peach']})
grade  =   pd.DataFrame({'Num':[123, 789, 1314], 
            'Grade':['A', 'B', 'C']})
```

```python
fruit
```

```python
grade
```

```python
pd.merge(fruit, grade, on='Num',how='left')
```

```python
pd.merge(fruit, grade,  on='Num',how='right')
```

```python
pd.merge(fruit, grade,  on='Num',how='inner')
```

```python
pd.merge(fruit, grade,  on='Num',how='outer')
```

```python
df1 = pd.DataFrame({'key': ['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df2 = pd.DataFrame({'key':['a','b','d'],
                   'data2': range(3)})
df1
```

```python
df2
```

```python
pd.merge(df1, df2) # how 안쓰면 자기가 알아서 공통된 항목으로 해줍니다. ㅎ
# 근데 보니까 HOW='inner' 인듯함.
```

```python
pd.merge(df1, df2, on='key')
```

```python
#merge는 교집합 결과를 반환한다.
df3 = pd.DataFrame({'lkey': ['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df4 = pd.DataFrame({'rkey':['a','b','d'],
                   'data2': range(3)})

df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
# 공통된 key 가 없었읍요
```

```python
 pd.merge(df1,df2, how='outer' )
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='outer')
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

다대다 병합은 잘 정의되어 있지만 직관적이진 않다.  
다대다 조인은 두 로우의 카르테시안 곱을 반환한다.

```python
pd.merge(df1, df2, how='right') #오른쪽 테이블에 존재하는 모든 키 조합을 사용한다.
#오른쪽에 해당하는 df2객체의 키값을 기준으로 병합, df1객체의 c는 빠지게 된다.
```
