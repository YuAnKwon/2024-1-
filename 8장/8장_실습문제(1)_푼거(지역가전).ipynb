1. data.columns.map(lambda x: x[:2]):  
- 이 부분은 모든 열 이름을 처리하여 각 열 이름의 처음 두 문자만 남기는 것입니다.
- map() 함수는 열 이름을 순회하면서 각각의 열 이름에 함수를 적용하고 그 결과를 반환합니다. 
- 여기서 사용된 람다 함수는 각 열 이름의 처음 두 문자를 추출합니다.

2. data.iloc[0]:  
- 이 부분은 데이터프레임의 첫 번째 행을 가져옵니다.  
- 따라서 이 코드는 데이터프레임의 첫 번째 행의 값들을 열 이름으로 설정합니다.  

3. data.columns = [data.columns.map(lambda x: x[:2]), data.iloc[0]]: 
- 이 부분은 열 이름을 변경합니다.
- 열 이름을 처음 두 문자로 바꾸는 부분과 데이터프레임의 첫 번째 행의 값을 열 이름으로 설정하는 부분을 조합하여 열 이름을 변경합니다.
- 즉, 각 열 이름의 처음 두 문자를 남기고 데이터프레임의 첫 번째 행의 값들로 열 이름을 대체합니다.

```python
data = pd.read_csv('지역가전.csv',encoding='UTF-8-SIG', header=1,skipfooter=2)
data = data.set_index(['지역','지역.1'])
data.index.names=['품목','상세품목']
data.columns.name='지역'
data.columns = [data.columns.str[:2], data.iloc[0]]
data.columns.names=['지역','지점']
data = data.iloc[1:]
data
```

### 3. 서울지역 지점 A, B, C 전체 컴퓨터 상세품목의 합을 구하시오. 

```python
data = data.astype("Int64")
a = data.loc['컴퓨터','서울'].cumsum(axis=1)
num = a.iloc[0,2] + a.iloc[1,2]
print(num)
```

```python
# 교수님 답
result = data.loc['컴퓨터','서울'].sum(axis=1)
result[0] + result[1]
```
