# stack과 unstack
- stack() 메서드:  
stack() 메서드는 열을 행으로 "스택(stack)"하여 다중 인덱스를 가진 Series(시리즈)를 반환합니다.
다중 인덱스의 가장 안쪽 수준의 열이 가장 안쪽 수준의 인덱스로 이동합니다.  
예를 들어, 열 인덱스를 행 인덱스로 이동하여 데이터프레임의 형태를 좁게 만듭니다.  
이 메서드는 데이터프레임의 열을 시리즈로 변환하여 다중 인덱스를 가진 시리즈를 반환합니다.
- unstack() 메서드:
unstack() 메서드는 행을 열로 "언스택(unstack)"하여 다중 인덱스를 가진 데이터프레임을 반환합니다.  
다중 인덱스의 가장 안쪽 수준의 인덱스가 가장 안쪽 수준의 열로 이동합니다.  
예를 들어, 행 인덱스를 열 인덱스로 이동하여 데이터프레임의 형태를 넓게 만듭니다.  
이 메서드는 데이터프레임의 인덱스를 열로 변환하여 다중 인덱스를 가진 데이터프레임을 반환합니다.

## 데이터 합치기
#### pandas 객체에 저장된 데이터는 여러가지 방법으로 합칠 수 있다. 
####     1. 'pandas.merge'는 하나 이상의 키를 기준으로 DataFrames의 row를 합친다. 
####     2. 'pandas.concat'는 하나의 축을 따라 객체를 이어 붙인다. 
####     3. 'combine_first' 메소드는 두 객체를 포개서 한 객체에서 누락된 데이터를 다른 객체에 있는 값으로 채울수 있게 한다. 

### 데이터베이스 스타일로 DataFrame 합치기 
#### 병합(Merge)과 조인(Join) 연산은 관계형 데이터베이스의 핵심이다. 
#### 하나 이상의 키를 사용해서 데이터 집합의 row를 합친다.  
#### pandas의 merge함수를 이용해서 데이터에 적용할 수있다.  

```python
df1 = pd.DataFrame({'key': ['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df2 = pd.DataFrame({'key':['a','b','d'],
                   'data2': range(3)})
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
#어떤 컬럼을 병합할지 명시하지 않으면, merge함수는 중복된 컬럼이름을 키로 사용
#(이 예에서는 'key')
#key 값이 공통적으로 가지고 있는 항목들만 병합
```

```python
pd.merge(df1,df2, on='key') # key를 공통으로 해주세요!
```

```python
#merge는 교집합 결과를 반환한다.
df3 = pd.DataFrame({'lkey': ['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df4 = pd.DataFrame({'rkey':['a','b','d'],
                   'data2': range(3)})

df3
```

```python
df4
```

```python
# 만약 두 객체에 중복된 컬럼이 하나도 없다면 따로 지정해주면 된다.
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
# lkey랑 rkey랑 둘다 공통인것만 나오게된다. default가 inner임.
```

pd.merge(df1, df2, how='outer') # 말그대로 합집합

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='outer')
```
