###  이걸로 출력된 값으로 encoding을 해야 불러올수있음.

```python
import chardet
import pandas as pd
filename = "수업자료/경상남도_창원시_공장등록현황_20210831.csv"
with open(filename, 'rb') as f:
    result = chardet.detect(f.readline())  # or read() if the file is small.
    print(result['encoding'])

```

```python
data = pd.read_csv('수업자료/경상남도_창원시_공장등록현황_20210831.csv',
                   encoding='UTF-8-SIG')
data
```
